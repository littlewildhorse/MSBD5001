```python
from sklearn.datasets import make_classification
from sklearn.linear_model import SGDClassifier
import pandas as pd
import numpy as np
import time
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
import numpy
seed = 7
numpy.random.seed(seed)
```

```python
trainFeatures=pd.read_csv('train.csv')#training data
TrainData=pd.read_csv('train.csv')#training data
testFeatures=pd.read_csv('test.csv')#testing data
heads=[row for row in trainFeatures]
label=trainFeatures['time'].values
```

```python
#根据参数构建一个特征
#从feature中提取参数通过classification来计算出real time
trainrealtime=[]
for i in range(0, 400):
    print(i)
    X1,Y1=make_classification(n_samples = TrainData.iat[i,7], n_features= TrainData.iat[i,8] , n_classes = TrainData.iat[i,9]  , n_clusters_per_class = TrainData.iat[i,10] ,n_informative = TrainData.iat[i,11], flip_y= TrainData.iat[i,12] , scale= TrainData.iat[i,13])
    clf = SGDClassifier(penalty = TrainData.iat[i,1], l1_ratio = TrainData.iat[i,2], alpha = TrainData.iat[i,3], max_iter = TrainData.iat[i,4], random_state = TrainData.iat[i,5], n_jobs = TrainData.iat[i,6])
    start = time.time()  
    clf.fit(X1, Y1)
    end=time.time()
    elapsed = (end - start)
    trainrealtime.append(elapsed)  
print(i,trainrealtime)
```

```python
#获取测试数据的实际时间
testrealtime=[]
for i in range(0, 100):
    print(i)
    X1,Y1=make_classification(n_samples = testFeatures.iat[i,7], n_features=testFeatures.iat[i,8] , n_classes = testFeatures.iat[i,9]  , n_clusters_per_class = testFeatures.iat[i,10] ,n_informative = testFeatures.iat[i,11], flip_y= testFeatures.iat[i,12] , scale=testFeatures.iat[i,13])
    import time
    from sklearn.linear_model import SGDClassifier
    clf = SGDClassifier(penalty = testFeatures.iat[i,1], l1_ratio =testFeatures.iat[i,2], alpha =testFeatures.iat[i,3], max_iter =testFeatures.iat[i,4], random_state =testFeatures.iat[i,5], n_jobs =testFeatures.iat[i,6])
    start = time.time()  
    clf.fit(X1, Y1)
    end=time.time()
    elapsed = (end - start)
    testrealtime.append(elapsed)    
testrealtime
```

```python
train=trainFeatures[heads[1:14]]
train=pd.get_dummies(train,columns=[heads[1]])
```

```python
test=testFeatures[heads[1:14]]
test=pd.get_dummies(test,columns=[heads[1]])
```

```python
#n_jobs的-1换成cpu的最大值16
train['n_jobs']=train['n_jobs'].replace(-1,16)
train['n_jobs']=train['n_jobs'].fillna(0)
test['n_jobs']=test['n_jobs'].replace(-1,16)
test['n_jobs']=test['n_jobs'].fillna(0)
```

```python
label=trainFeatures[heads[-1]]
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
import numpy
seed = 7
numpy.random.seed(seed)
```

```python
#建立全连接的神经网络
model = Sequential()
model.add(Dense(output_dim=800, input_dim=17, init='uniform', activation='relu'))
#加入dropout层
model.add(Dropout(0.2))
model.add(Dense(output_dim=1, input_dim=800,init='uniform'))
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
#对数据进行标准化
train=train.apply(lambda x: (x - np.min(x)) / (np.max(x) - np.min(x))) 
test=test.apply(lambda x: (x - np.min(x)) / (np.max(x) - np.min(x))) 

#这里如果能同一度量进行标准化效果可能会更好
#因为实际差距train和test的差距并不是很大，所以分开标准化不会有太大影响
```

```python
#加入实际时间作为feature
realtime=pd.DataFrame(trainrealtime,columns=['realtime'])
train['realtime']=realtime
testrealtime=pd.DataFrame(testrealtime,columns=['realtime'])
test['realtime']=testrealtime
```

```python
#拟合模型并迭代1000次
model.fit(train, label, nb_epoch=1000, batch_size=5,  verbose=2)
# evaluate the model
```

```python
#对test进行预测
y1=model.predict(test)
```

```python
result=[]
for i in y1:
    result.append(i[0])
```

```python
#将小于0的值转换成realtime的值
sum=0
n=0
y1=result
for i in range(0,100):
    if y1[i]<0:
        print(i,end=' ')
        sum=sum+label[i]
        n=n+1
        print(y1[i],end=' ')
        print(label[i])
        y1[i]=testrealtime['realtime'][i]
```

```python
#将文件写入到csv中并保存
submission=pd.DataFrame({'id':testFeatures['id'],'time':y1})
submission.to_csv('predictionresult.csv',columns=['id','time'],index=None)
```
